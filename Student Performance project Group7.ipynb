Background of the Project
        Your best friend is an administrator at a large school. The school makes every student take year-end math, reading, and writing exams. Since you have recently learned data manipulation and visualization, you suggest helping your friend analyze the score results. The school's principal wants to know if test preparation courses are helpful. She also wants to explore the effect of parental education level on test scores.
        
        The data
The file has the following fields (source):
"gender" - male / female
"race/ethnicity" - one of 5 combinations of race/ethnicity
"parent_education_level" - highest education level of either parent
"lunch" - whether the student receives free/reduced or standard lunch
"test_prep_course" - whether the student took the test preparation course
"math" - exam score in math
"reading" - exam score in reading
"writing" - exam score in writing

```python
reading_avg = std_perf.groupby('test_prep_course')[['reading']].mean()
reading_avg
```

```python
r =reading_avg.reading.plot(x ='test_prep_course', kind = 'bar' ,  rot = 0)
r.set(title = 'The Average reading scores of students with or without the test preparation course');
#r.set_xticklabels(r.get_xticklabels());
```

```python
parent_edulevel_avg = std_perf.groupby('parent_education_level')[['reading','writing','math']].mean()
parent_edulevel_avg
```

```python
m =parent_edulevel_avg.plot( kind = 'bar',figsize=(20, 10) ,  rot = -10)
m.set(title = "Influence of parent's education level on the average scores of the students");
m.set_xlabel("Parent's educational levels")
m.set_ylabel("Average Scores of students")
```

```python
m =parent_edulevel_avg.plot( kind = 'line',figsize=(20, 10) ,  rot = -10)
m.set(title = "The Average scores of students based on their parent's education level");
m.set_xlabel("Parent's educational levels")
m.set_ylabel("Average Scores of students")
```

```python

score_based = parent_edulevel_avg.T 
score_based
```

```python
n =score_based.plot( kind = 'line',figsize=(20, 10) ,  rot = -10)
n.set(title = " Subject based performance considering the parent's education level");
n.set_xlabel(" Subject Tested ")
n.set_ylabel("Average Scores of students")
```
